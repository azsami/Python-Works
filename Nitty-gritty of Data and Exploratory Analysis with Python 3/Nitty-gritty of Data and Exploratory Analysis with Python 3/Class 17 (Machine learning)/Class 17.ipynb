Let's assume our hypothesis function for our regression line is <br> <br> 
$ h_{0}(\theta) = \theta_{0} + \theta_{1}\times x $ <br> <br>
Here, $ h_{0}(\theta)$ is <b> y </b> <br>
     $ \theta_{1} $ is <b>m</b>(Slope) and <br>
     $ \theta_{0} $ is <b>b</b>(y-intercent) <br><br>
     
Now our main objective is to find such $ \theta_{1} $ and $ \theta_{0} $ so that our loss function or cost function will be minimum, <br><br>

Formula of Cost function -- <br><br>
$J (\theta_{0}, \theta_{1}) = \frac{1}{n} \sum_{i=1}^{n} \left (Y_{orginal} - Y_{predict} \right )^{2}$<br><br>
$=> J (\theta_{0}, \theta_{1}) = \frac{1}{n} \sum_{i=1}^{n} \left (Y_{i} -  h_{0}(\theta) \right )^{2}$<br><br>
$=> J (\theta_{0}, \theta_{1}) = \frac{1}{n} \sum_{i=1}^{n} \left (Y_{i} -  ( \theta_{0} + \theta_{1}\times X_{i}) \right )^{2}$ <br><br>
Now to find the minimum $\theta_{0}$ and $\theta_{1}$ we will use <b>Gredient Decesnt</b> algorithm - <br><br>

Repeat until convergence {

$ \theta_{j} := \theta_{j} - \alpha \frac{\delta}{\delta \theta_{j}} J(\theta_{j}) $

}

Minimizing $\theta_{0} $ or b : <br>
<br><br>
Minimizing $\theta_{1} $ or m : <br>

Understanding partial derivaties <br>

So, Cost function - <br>
$J (\theta_{0}, \theta_{1}) = \frac{1}{n} \sum_{i=1}^{n} \left (Y_{i} -  ( \theta_{0} + \theta_{1}\times X_{i}) \right )^{2}$ <br><br>
Partial derivatives of Cost function with respect to $\theta_{0}$ - <br>
$ \frac{\delta{J (\theta_{0}}, \theta_{1})}{\delta\theta_{0}} = \frac{2}{n} \sum_{i=1}^{n} -\left (Y_{i} -  ( \theta_{0} + \theta_{1}\times X_{i}) \right )$ <br><br>
Partial derivatives of Cost function with respect to $\theta_{1}$ - <br>
$ \frac{\delta{J (\theta_{0}, \theta_{1})}}{\delta\theta_{1}} = \frac{2}{n} \sum_{i=1}^{n} -X_{i}\left (Y_{i} -  ( \theta_{0} + \theta_{1}\times X_{i}) \right )$ <br><br>

So new $\theta_{0} = \theta_{0} - \alpha\frac{\delta{J (\theta_{0}}, \theta_{1})}{\delta\theta_{0}}$<br><br>
and new $\theta_{1} = \theta_{1} - \alpha\frac{\delta{J (\theta_{0}}, \theta_{1})}{\delta\theta_{1}}$<br><br>

This process will repeat untill we get the minimum $\theta_{0}$ and $\theta_{1}$


```python
import math

def gradient_descent(x, y):
    
    Theta_0 = Theta_1 = 0
    n = len(x)
    learning_rate = .001
    iteration = 100000
    
    cost_prev = -100000000000000000000000
    
    for i in range(iteration):
        y_pred = Theta_0 + Theta_1*x
        
        cost = (1/n) * sum(pd.Series(y-y_pred).apply(lambda a : a**2))
        
        Partial_derivatives_respect_to_Theta_0 = - (2/n) * sum(y-y_pred)
        
        Partial_derivatives_respect_to_Theta_1 = - (2/n) * sum(x *(y-y_pred))
        
        Theta_0_prev = Theta_0
        Theta_1_prev = Theta_1
        
        Theta_0 = Theta_0 -  learning_rate * Partial_derivatives_respect_to_Theta_0
        Theta_1 = Theta_1 -  learning_rate * Partial_derivatives_respect_to_Theta_1
     
    
        print('Theta_0 {}, Theta_1 {}, cost {} iteration {}'.format(Theta_0_prev, Theta_1_prev, cost, i))
        
        
        if(math.isclose(cost, cost_prev, rel_tol=1e-15)):
            break
            
        cost_prev = cost
        
    return(Theta_0_prev, Theta_1_prev)

(Theta_0, Theta_1) =gradient_descent(x, y)

```
