```python
from urllib.request import urlopen
import time


Check_next_page = 1

i = 1

with open("Qoutes.csv", "w") as f:
    f.write('Author,Tags,Quotes\n')
    
    
    while Check_next_page == 1:
        
        
        page_url = "http://spidyquotes.herokuapp.com/page/{}/".format(str(i))
        
        
        print(page_url)


        with urlopen(page_url, timeout = 100) as m:
            page_soup = soup(m.read(), "html.parser")
        
        for p_soup in page_soup.findAll("div", {"class" : "quote"}):
            
            Author = p_soup.small.text

            tags = ''

            tag_soup = p_soup.findAll('a', {'class' : 'tag'})

            j = 1

            for tag in tag_soup:
                if j == len(tag_soup):
                    tags = tags + tag.text
                else:
                    tags = tags + tag.text + " | "

                j = j+1


            Quotes = p_soup.span.text.replace(',', ' ')

            try:
                f.write(Author + ',' + tags + ',' + Quotes + '\n')
            except:
                pass
    
            
        if len(page_soup.findAll("li", {"class" : "next"})) == 0:
            
            Check_next_page = 0
                
                
        i+=1
        
        time.sleep(1)

print("Done")
```
