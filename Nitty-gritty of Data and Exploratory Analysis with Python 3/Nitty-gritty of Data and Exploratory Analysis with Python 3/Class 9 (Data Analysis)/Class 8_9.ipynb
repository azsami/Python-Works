### Filtering DataFrame by Boolean indexing

#### Find all Athlete's records who have won gold medals.

```python
oo.head() 
```

```python
oo[ oo['Medal'] == 'Gold' ]
```

#### Find all women Athlete's records who have won silver medals.

```python
# and or >= <=

# &  |  == >= <=


    

oo[ (oo['Medal'] == 'Silver')  &   (oo['Gender'] == 'Women') ]
```

#### Find all records of the femous USA's woman athlete named Florence.

```python
oo[( oo['Athlete'].str.contains('Florence') )  &  (oo['Gender'] == 'Women')  &  (oo['NOC'] == 'USA')]

# oo[(oo['NOC'] == 'USA') & (oo['Gender'] == 'Women') & (oo['Athlete'].str.contains('Florence'))]

# oo[(oo['Athlete'].str.contains('lor'))]
```

#### Find all records of athletes who have won the men's gold medals in singles badminton over the history of Olympics? Sort the results alphabetically by the player's  names.

```python
gmb = oo[ (oo['Gender'] == 'Men') &  (oo['Sport'] == 'Badminton') & (oo['Event'] == 'singles') & (oo['Medal'] == 'Gold') ]

gmb.sort_values(by = 'Athlete')
```

#### Which three countries have won the most medals from 1984 to 2008 ?

```python
 >= 1984       <=2008
```

```python
med = oo[(oo['Edition'] >= 1984)  &  (oo['Edition'] <= 2008)]

med['NOC'].value_counts(ascending = True).head(10)
```

#### Find all Records of the male gold medal winners for the 100m event over the history of olympics. List the results starting with the most recent. Show the Olympic City, Edition, Athlete and the country they represent.

```python
oo.head()
```

```python
a = oo[ (oo['Gender'] == 'Men') & (oo['Medal'] == 'Gold') & (oo['Event'] == '100m')]
b = a.sort_values(by = 'Edition', ascending = False)
b[['City', 'Edition', 'Athlete', 'NOC']]
```
