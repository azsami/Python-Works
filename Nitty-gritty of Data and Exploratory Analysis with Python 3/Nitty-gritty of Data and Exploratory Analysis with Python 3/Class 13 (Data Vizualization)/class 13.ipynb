# Basic plotting

```python
import pandas as pd

import matplotlib.pyplot as plt

%matplotlib inline
```

```python
sales = [100, 120, 340, 200, 250, 234, 400]

days = list(range(1,8))
```

```python
plt.plot(days, sales)

# plt.barh(days, sales)

plt.show()
```

```python
plt.plot(days, sales, label = 'Sales Amount')

plt.xlabel('No. of Days')
plt.ylabel('Sales')
plt.title('Total Sales on Daily Basis')
plt.legend(loc = 'best')

# 'upper left', 'upper right', 'lower left', 'lower right'``

plt.show()
```

```python
oo = pd.read_csv('olympics.csv', skiprows = 4)
oo.head()
```

## Plot types

### What were the different sports in the first olympics? Plot them using different graphs.

```python
fo = oo[oo['Edition'] == 1896]
```

```python
fo['Sport'].value_counts().plot();
```

### Line plot

```python
fo['Sport'].value_counts().plot(kind = 'line');
```

### Bar plot

```python
fo['Sport'].value_counts().plot(kind = 'bar');
```

### Horizontal bar plot

```python
fo['Sport'].value_counts().plot(kind = 'barh');
```

### Pie chart

```python
fo['Sport'].value_counts().plot(kind = 'pie');
```

```python
us = oo[oo['NOC'] == 'USA']
```

```python
us.Medal.value_counts().plot(kind = 'line');
plt.title('USA medals');
```

## Plot colors

```python
fo['Sport'].value_counts().plot(kind = 'line', color = 'darkorange')
```

```python
fo['Sport'].value_counts().plot(kind = 'bar', color = 'maroon');
```

## figsize()

```python
fo['Sport'].value_counts().plot(kind = 'line', color = 'darkorange', figsize = (10,3));
```

## Colormaps

```python
fo['Sport'].value_counts().plot(kind = 'pie', colormap = 'Pastel1');
```

### How many medals have been won by men and women in the history of the Olympics. How many gold, silver and bronze medals were won for each gender?

```python
sns.countplot(x = 'Medal', data = oo, hue = 'Gender', order = ['Bronze', 'Gold', 'Silver'], hue_order = ['Women', 'Men']);

```

## Exercise

### Plot the number of medals achieved by the Chinese team (men and women) in Beijing 2008 using:
- matplotlib
- Seaborn

```python
oo.groupby(['Gender','Medal']).count()['City'].plot(kind = 'bar', stacked = True);
```

```python
CHN_2008 = oo[(oo['NOC'] == 'CHN') & (oo['Edition'] == 2008)]
```

```python
CHN_2008.Gender.value_counts().plot(kind = 'bar', stacked = True);
```

```python
sns.countplot(data = CHN_2008, x = 'Gender');
```

### How can you use colormaps to give the data more meaning?

https://matplotlib.org/_images/colormaps_reference_03.png

```python
sns.countplot(data = CHN_2008, x = 'Gender', palette = 'bwr');
```
