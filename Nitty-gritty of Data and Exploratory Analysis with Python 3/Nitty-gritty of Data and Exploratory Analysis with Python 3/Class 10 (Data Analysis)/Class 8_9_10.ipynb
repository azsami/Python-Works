# Groupby

```python
df = pd.DataFrame({ 'Brand' : ['MS','MS','Google','Google','Yahoo','Yahoo'],
                    'Sales Person' : ['Robin','Adam','Sarah','Adrew','Rahim','Karim'],
                    'Sales' : [1000, 2000, 3000, 200, 3000, 200],
    
})

df
```

```python
df.groupby('Brand')
```

```python
for group_key, group_value in df.groupby('Brand'):
    print(group_key)
    print(group_value)
```

```python
df.groupby('Brand').count()
```

```python
df.groupby('Brand').agg({'Sales' : ['count', 'min', 'max']})
```

```python
df.groupby('Brand').agg({'Sales' : 'sum'})
```

### Find Athlete named LEWIS, Carl 's last and latest medal winning year and his total number of medals

```python
oo[oo['Athlete'] == 'LEWIS, Carl'].groupby('Athlete').agg({'Edition' : ['min', 'max', 'count']})
```

### Create a list showing the total number of medals won for each country over the history of the Olympics. For each country, include the year of the first and most recent Olympic medal wins.

```python
oo.groupby('NOC').agg({'Edition' : ['min', 'max', 'count']})['Edition'].sort_values(by = 'count', ascending = False)
```
