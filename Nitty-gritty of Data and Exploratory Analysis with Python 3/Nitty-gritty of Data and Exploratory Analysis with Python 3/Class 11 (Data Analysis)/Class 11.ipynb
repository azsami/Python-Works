```python
df.dropna(how = 'all') 
```

```python
train = pd.read_csv('http://bit.ly/kaggletrain')
train.head()
```

```python
train['Name_length'] = train['Name'].apply(len)
train.head()
```

```python
train[['Name', 'Name_length']].head(10)
```

```python
def get_first_name(string):
    return string.split(',')[0]

train['Name'].apply(get_first_name)
```

```python
train['Name'].apply(lambda x : x.split(',')[0])
```

```python
train[train['Sex'].apply(lambda x : x == 'female')]

# train[train['Sex'] == 'female']
```
