```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

%matplotlib inline
```

```python
a = np.linspace(0,5,11)
b = a ** 2
```

```python
plt.plot(a,b, color = 'r')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.title('This is a chart')
```

### Subplot

```python
# plt.subplot(nrows, ncols, plot_number)

plt.subplot(1,2,1)
plt.plot(a,b,'r', ls = '--')


plt.subplot(1,2,2)
plt.plot(b,a,'g--', lw = 3)
```

### Matplotlib OOM

```python
fig = plt.figure()

axes = fig.add_axes([.1, .1, .8, .8])  # left bottom right top (range 0 to 1)

axes.plot(a,b, 'r--', lw = 3)
axes.set_xlabel('X Axis')
axes.set_ylabel('Y Axis')
axes.set_title('This is a chart')
```

```python
fig = plt.figure()

ax1 = fig.add_axes([.1, .1, .8, .8])
ax2 = fig.add_axes([.2, .5, .4, .3])


ax1.plot(a,b, 'r--')
ax1.set_xlabel('X Axis for Axis 1')
ax1.set_ylabel('Y Axis for Axis 1')


ax2.plot(b,a, 'g--')
ax2.set_xlabel('X Axis for Axis 2')
ax2.set_ylabel('Y Axis for Axis 2')
```

```python
fig = plt.figure(figsize = (12,5))

ax1 = fig.add_axes([.1, .1, .8, .8])
ax2 = fig.add_axes([.2, .5, .4, .3])


ax1.plot(a,b, 'r--')
ax1.set_xlabel('X Axis for Axis 1')
ax1.set_ylabel('Y Axis for Axis 1')


ax2.plot(b,a, 'g--')
ax2.set_xlabel('X Axis for Axis 2')
ax2.set_ylabel('Y Axis for Axis 2')
```

```python
fig, axes = plt.subplots(1,3, figsize = (12,5))

axes[0].plot(a, a**2, 'r--',   a, a**3, 'g*--')
axes[0].set_title('Plot 1')

axes[1].plot(a, a**3, 'g*--')
axes[1].set_title('Plot 2')

axes[2].plot(a, b)
axes[2].set_title('Plot 3')
```

# Distribution Plots
* disjointplot
* joinplot
* pairplot
* rugplot
* kdeplot

```python
tips = sns.load_dataset('tips')

tips.head()
```

### Disjointplot

```python
sns.distplot(tips['total_bill'], kde = False)
```

### Joint plot
Kind :
* scatter
* reg
* resid
* kde
* hex

```python
sns.jointplot(x = 'total_bill', y= 'tip', data = tips, kind = 'scatter')
```

```python
sns.jointplot(x = 'total_bill', y= 'tip', data = tips, kind = 'hex')
```

```python
sns.jointplot(x = 'total_bill', y= 'tip', data = tips, kind = 'reg')
```

```python
sns.jointplot(x = 'total_bill', y= 'tip', data = tips, kind = 'kde')
```

```python
sns.jointplot(x = 'total_bill', y= 'tip', data = tips, kind = 'resid')
```

### Pairplot
* Will plot pair wise relationship across entire dataframe (for numerical columns)
* To see categorical columns need to use hue

```python
tips.head()
```

```python
sns.pairplot(tips)
```

```python
sns.pairplot(tips, hue = 'sex')
```

### Rugplot & Kdeplot

```python
sns.rugplot(tips['total_bill'])
```

```python
sns.kdeplot(tips['total_bill'])
```

```python
sns.kdeplot(tips['total_bill'])
sns.rugplot(tips['total_bill'])
```
